### Gráficos Básicos

```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df= pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
df
```

### Gráfico de barras

```python
df_agrupado= df.groupby('day', sort=False).agg({'total_bill': np.sum}).reset_index() #agrupando o dataframe
df_agrupado
```

```python
px.bar(df_agrupado, x='day', y='total_bill') #gráfico de barras
```

```python
df_agrupado_genero= df.groupby(['day', 'sex'], sort=False).agg({'total_bill': np.sum}).reset_index()
df_agrupado_genero
```

```python
px.bar(df_agrupado_genero, x='day', y='total_bill', color='sex') #a divisão de cores é feita pelo genero.
```

```python
px.bar(df_agrupado_genero, x='day', y='total_bill', color='sex', barmode= 'group') #barmode separa os gráficos para não ficarem empilhados
```

### Gráfico de dispersão

```python
px.scatter(df, x='total_bill', y='tip', color='day', hover_name='time') #hover_name destaca a info desejada
#quanto maior a conta, maior o aumento da gorjeta
```

### Gráfico de pizza

```python
pedidos_por_dia= df['day'].value_counts() #quantos pedidos foram feitos a cada dia
#repare que é uma série
```

```python
pedidos_por_dia.index
pedidos_por_dia.values
```

```python
px.pie(pedidos_por_dia, names= pedidos_por_dia.index, values=pedidos_por_dia.values)
```

### Gráficos estatísticos

```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df= pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
```

```python
df.head()
```

### Histograma
é a frequencia que um mesmo valor aparece

```python
px.histogram(df, x='total_bill', color='sex', nbins=40, histnorm='probability density', marginal='rug', title= 'Histograma da taxa')
#bins mostra o intervalo das barras, como serão mostradas. marginal rug mostra aonde há maior concentração de dados, conforme os riscos mostrados acima (ocorrência).
```

### Gráfico de distribuição
mostra a densidade de dados

```python
import plotly.figure_factory as ff
import numpy as np
```

```python
x1= np.random.randn(1000) #gerando 1000 valores que seguem uma distribuição normal
x2= np.random.randn(1000) +5
```

```python
ff.create_distplot([x1, x2], group_labels=['Normal 1', 'Normal 2']) #plotando os 2 gráficos
```
